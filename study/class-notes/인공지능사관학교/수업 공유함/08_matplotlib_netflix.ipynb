```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
netflix = pd.read_csv('./netflix_titles.csv')
netflix
```

```python
netflix.head(1)
netflix.info()
```

```python
# 결측치가 많이 보인다 
for i in netflix.columns:
    missingValueRate = netflix[i].isna().sum() / len(netflix) * 100
    if missingValueRate > 0:
        print("{} null rate: {}%".format(i, round(missingValueRate, 2)))
```

```python
# 결측치 전부다 교체해주기
netflix['country'] = netflix['country'].fillna('No Data')
netflix['director'] = netflix['director'].fillna('No Data')
netflix['cast'] = netflix['cast'].fillna('No Data')
netflix.head()
```

```python
netflix.dropna(axis=0, inplace=True)
netflix.info()
```

```python
netflix.isna().sum()
```

```python
# 파생 컬럼 추가 하기 기존의 존재하는 변수를 활용하여 새로운 정보를 추가로 생성하는 과정 << 피처 엔지니어링 중요?
# rating << 등급이 굉장히 결이 다르다 그래서 교수님이 우리만에 시청 등급을 만들도록 하겠습니다
netflix['age_group'] = netflix['rating']
netflix.head()
```

```python
# age_group dic 라는 사전을 보고 
age_group_dic = {
    'G': 'All',
    'TV-G': 'All',
    'TV-Y': 'All',
    'PG': 'Older Kids',
    'TV-Y7': 'Older Kids',
    'TV-Y7-FV': 'Older Kids',
    'TV-PG': 'Older Kids',
    'PG-13': 'Teens',
    'TV-14': 'Young Adults',
    'NC-17': 'Adults',
    'NR': 'Adults',
    'UR': 'Adults',
    'R': 'Adults',
    'TV-MA': 'Adults'
    }

# 각각의 요소를 제어할때 .map 사용 <<< 중요
netflix['age_group'] = netflix['age_group'].map(age_group_dic)
netflix.head(2)
```

```python
netflix.to_csv('netflix_preprocessed.csv', index=False)
```
