```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3,4], [1,4,9,16], label='Square', marker="o")

# 이름 정해주기
plt.xlabel('X-Label')
plt.ylabel('Y-Label')

#범위 정해주기
plt.xlim([0,5])
plt.ylim([0,15])

#라벨 사용학
plt.legend()
plt.show()
```

```python
# x축 값
years = ['2030','2040','2050']
plt.xlabel('Years')
# y축 값
values = [300, 100, 700]
plt.ylabel('Values')

# 막대 그래프 bar 만들기
plt.bar(years,values, color = ['r','g','b'], label=['legend','legend2','legend3'])
plt.legend()
plt.show
```

```python
categories = ['A','B','C','D']
values = [25,40,30,20]
plt.ylim([0,50])
plt.bar(categories, values, color='pink')

# 밸류즈 안에 데이터 값이 4개가 있다 0에 해당하는 25, 1에 해당하는 40 .... 가 enumerate이다
for i, value in enumerate(values):
    # plt.text(x 좌표, y 좌표, 숫자화 시켜주기, )
    plt.text(i, value+1, str(value), ha='center', va='bottom', color='black', fontsize=10)

plt.title('KOSA chart')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```
