```python
import numpy as np
```

```python
cus1 = np.array([150,25,0])
cus2 = np.array([200,30,0])
cus3 = np.array([50,10,1])
cus4 = np.array([300,45,0])
customer1 = np.vstack((cus1,cus2,cus3,cus4))
print(customer1)
```

```python
cus5 = np.array([120,20,0])
cus6 = np.array([180,35,1])
cus7 = np.array([75,5,0])
cus8 = np.array([160,25,0])
customer2 = np.vstack((cus5,cus6,cus7,cus8))

data_customer = np.vstack((customer1, customer2))
print(data_customer)
```

```python
customer_sum = data_customer.sum(0)
print(customer_sum)
average_sum = data_customer.mean(0)
print(average_sum)
combined_data = np.vstack((customer_sum,average_sum))

print(f'댓글 길이 총합 : {combined_data[0,0]}')
print(f'좋아요 수 총합 : {combined_data[0,1]}')
print(f'댓글 길이 평균 : {combined_data[1,0]}')
print(f'좋아요 수 평균 : {combined_data[1,1]}')


```
