```python
import pandas as pd
import numpy as np
```

```python
customer_dic = {
'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace','Hyemin'],
'comment_length': [150, 200, 50, 300, 120, 180, 75, 160],
'likes': [25, 30, 10, 45, 20, 35, 5, 28],
'is_spam': [False, False, True, np.nan, False, True, np.nan, False],
'has_image': [True, False, True, True, False, False, True, True]
}
```

```python
# 1번 customer_dic를 데이터프레임으로 만들어주기
customer = pd.DataFrame(customer_dic)
customer
```

```python
# 2번 is_spam 결측치 확인하기 
#is_spam 결측치를 False로 대체하기
customer['is_spam'] = customer['is_spam'].fillna('False')
customer['is_spam'] = customer['is_spam'].replace('False', False)
customer.info()

customer
```

```python
# 3번 has_image 열을 기준으로 이미지가 포함된 댓글과 포함되지 않은 댓글의 평균 likes를 구하세요 
customer.groupby('has_image')['likes'].mean()
```

```python
# 4번 상품을 받을 고객의 이름을 구하세요 조건은 아래와 같습니다
customer.query('comment_length >= 100 & likes >= 20 & is_spam == False & has_image == True')
```

```python
# 5번 고객 데이터에서 이름만 추출하여 시리즈 형태로 생성하고 하위 3줄의 데이터를 출력세요
pd.Series(customer['name']).tail(3)
```

```python
# 6번 고객 데이터를 to_csv() 함수를 사용하여 'customer.csv' 파일로 저장하세요
customer.to_csv('customer.csv', index=False)
```
