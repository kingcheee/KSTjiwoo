```python
import pandas as pd
```

```python
netflix = pd.read_csv('./netflix_cleaned.csv')
netflix
```

```python
# 통계 구하기 (mean, median, sum, min, max, var << 분산, 표준편차) 
print(f'평균 상영시간={netflix['duration'].mean()}')
```

```python
# type을 보면 값이  tv show 도 있고 movie 도 있고 값이 어떤게 들어가져잇는지 알고싶을때가 있다 
print(f'count={netflix['type'].count()}')
print(f'value_counts= {netflix['type'].value_counts()}')
```

```python
# 원하는 통계만 보고 싶을때 agg()를 사용한다 통계량을 추측하고 싶을때
print(netflix.agg({'release_year' : ['min', 'max', 'median', 'std'],'duration' : ['min', 'max', 'median', 'std']}))
```

```python
# 집합별 통계 구하기
#groupby(원하는 집합)[기준 열]
netflix.groupby('type')['duration'].mean()
```

```python
netflix.groupby(['type', 'country'])['duration'].mean()
```
