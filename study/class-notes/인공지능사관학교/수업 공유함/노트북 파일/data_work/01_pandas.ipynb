```python
# 특정 열을 선택하는 목적이 있으면 마찬가지로 특정 행을 선택도 해보고싶다
# SQL의 WHERE절을 데이터프레임으로 이용해서 뽑는단 마인드
netflix = pd.read_csv('./netflix_selena.csv')
netflix['release_year'] > 2015
```

```python
# 전체를 괄호로 묶어주고 물결표시를 하면 not이라는 뜻이된다
# 범위에 괄호를 잘 설정해주어야한다 
less2015 = netflix[~(netflix['release_year'] > 2015)]
less2015.head()
```

```python
# 결측치 처리 (대체)
# fillna << 결측치만 찾아서 집어넣는다
netflix['country'] = netflix['country'].fillna('No Data')
netflix
```

```python
# 결측치 뿐만이 아니다 다른것도 변경가능
netflix['director'] = netflix['director'].replace(np.nan, '박규연')
netflix
```

```python
# .dropna() 값을 대체하는게 아니라 결측치 처리하는 값 - 해당하는 열을 제거해버리는것이다 
# 방향만 나타낼때는 1이 열이다 연산할때는 0이 열이다
# inplace=True 는 원본자체를 없애주고 inplace=False는 그게 아니다 <<<   
netflix.dropna(axis=1, inplace=True)
netflix
```

```python
netflix = pd.read_csv("./netflix_selena.csv")
netflix
```

```python
#결측치 처리(열 제거) 디렉터에 널값이면 제거하자
# subset = 열을 설정해줘서 그것중에서 dropna << nan 이면 행을 죽이겠다
netflix.dropna(subset=['director'], inplace=True)
netflix
```

```python
# 결측치 처리 (특정 열 제거)
if netflix['listed_in'].isna().any():
    netflix.drop(columns=['listed_in'], inplace=True)
netflix
```
