# Numpy 배열 기초

```python
#리스트가 아닌 Numpy 배열은 np.array() 함수를 통하여 생성할 수 있다

import numpy as np

# np.array 라는 객체 타입으로 생성이 되었다는걸 알 수 있다
arr = np.array([1,2,3])
print(type(arr))
```

```python
list1 = [1,2,3,4,5]
list2 = [6,7,8,9,10]

# 파이썬 기본 리스트에서 +를 사용하면 그냥 이어붙인다
list1 + list2
```

```python
arr1 = np.array(list1)
arr2 = np.array(list2)
# 이걸 통해서 행열의 덧셈 혹은 곱셈을 할 수 있다
arr1 + arr2
```

```python
# 0차원 배열 - 스칼라
arr0d = np.array(1)
print("0차원 배열:", arr0d)

# 1차원 배열 - 백터
arr1d = np.array([1, 2, 3, 4, 5])
print("1차원 배열:", arr1d)

# 2차원 배열 - 메트릭스
arr2d = np.array([[1, 2, 3], [4, 5, 6]])
print("2차원 배열:", arr2d)

# 3차원 이상의 배열 - 텐서 << 인공지능은 결국 행열의 흐름이다
arr3d = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])
print("3차원 배열:", arr3d)
```

```python
#0차원
a = np.array(1)
# 차원을 출력해주는 코드
a.ndim
#1 차원
# 결국 대괄호의 갯수 n이 n차원 인거다
a = np.array([1,2,3])
a.ndim
#2차원
a = np.array([[1,2,3],[4,5,6]])
a.ndim
#3차원
a = np.array([[[1,2,3],[4,5,6]],[[1,2,3],[4,5,6]]])
a.ndim
```

```python
#괄호 안에 괄호가 들어가주어야 한다
np.zeros((3,3))
```

```python
# 첫번째 대괄호에서 3개가 생성이 되었고
# 두번째 대괄호에서 4개가 생성이 되었고
# 세번째 대괄호에서 2개가 생성이 되었다

np.zeros((3,4,2))
```

```python
# 1로 이루어진 배열
np.ones((2,3))
```

```python
np.ones((2,3,4))
# 어떠한 형태로 만들어지는 3차원까지는 머릿속으로 그려져야 한다
```

```python
# 저 배열을 특정 숫자로 채우고 싶을때
np.full((3,3), 10)
```

```python
# 단위 행열은 행열의 곱셈을 했을때 자기자신이 나오도록 하는것
# 대각선이 1이고 나머지는 0인 행렬을 뜻한다
np.eye(3)
```

```python
# 범위배열 파이썬에서 range() 라는것을 했다
list(range(10))
# numpy는 선형대수, 소숫점으로 되어있는것도 많이 사용한다
```

```python
# 0.5 씩 건너뛰게 해줄 수 있고, 2 씩 건너뛰게 해줄 수 있다 다 된다
np.arange(0, 10, 0.5)
```

```python
# np.arange(시작,끝,간격) 
np.arange(0, 10, 2)
type(np.arange(0, 10, 0.5))
```

```python
#np.linspace() 는 시작점과 끝점 사이를 균등하게 나눈다
np.linspace(0, 1, 5)
```
